```python
from ultralytics import RTDETR

model = RTDETR("rtdetr-l.pt")

results = model.train(
    data='./data.yaml',
    epochs=100,
    imgsz=768,
    batch=8,
    device='cuda',
    workers=4,
    patience=40,
    cos_lr=True,
    project='experimentos_wafer',
    name='rtdetr_wafer',
    # cache=True,
)
```

```python
model = RTDETR('./runs/detect/experimentos_wafer/rtdetr_wafer/weights/best.pt')

metricas = model.val(
    data='./data.yaml',
    split='test',
    project='experimentos_wafer',
    name='teste_rtdetr',
)

print(f"mAP50 Box: {metricas.box.map50:.3f}")
print(f"mAP50-95 Box: {metricas.box.map:.3f}")
```

```python
print(f"Precision: {metricas.box.p:.3f}")
print(f"Recall: {metricas.box.r:.3f}")
print(f"F1-Score: {metricas.box.f1:.3f}")
print(f"mAP50: {metricas.box.map50:.3f}")
print(f"mAP50-95: {metricas.box.map:.3f}")
```
